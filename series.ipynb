## PANDAS SERIES

##### A series is 1-d labelled array capable of holding any data type. The axis labels are collectively called the axis labels

### CREATING SERIES

##### There are multiple ways of creating a pandas series:

```python
import numpy as np
import pandas as pd
```

```python
# labels are indexes in pandas

labels = ["a","b","c"]
my_list = [10,20,40]
my_array = np.array([10,20,30])
d = {1:10, 2:20, 3:30}
```

```python
l = pd.Series(my_list)
li = pd.Series(my_list, index=labels)
print("list: \n",l)
print("list with index a b c: \n",li)
```

```python
pd.Series(my_array, index=labels)
```

```python
pd.Series(d)
```
