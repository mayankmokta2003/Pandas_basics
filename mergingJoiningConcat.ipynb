```python
import numpy as np 
import pandas as pd
```

### Merging 2 dataframes

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
# here it automatically takes on = "employee_id" as its common and merges on the basis of employee_id
pd.merge(employees,salaries)
```

```python
# how="inner" means merge happens on the basis of common parts like eid is 1,2,3 common 
pd.merge(employees,salaries, on="employee_id", how="inner")
```

```python
# how="outer" means dono table ka sara data lao based on employee_id
pd.merge(employees,salaries, on="employee_id", how="outer")
```

```python
# how="left" means left table ka sara data rakho or right table ka sirf matching data add kroo
pd.merge(employees,salaries,on="employee_id",how="left")
```

```python
# how="right" means right table ka sara data rakho or left table ka matching data add karo
pd.merge(employees,salaries,on="employee_id",how="right")
```

### Concatination of 2 dataframes

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
# here axis is automatically 0 so concatination happens on the basis of columns
pd.concat([df1,df2])
```

```python
# concatination happens on the basis of columns
pd.concat([df1,df2],axis=1)
```

### Joining 2 dataframes

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how="outer")
```

```python
df2.join(df1)
```
