### Creating a Data Frame

```python
data = {
    "Name": ["John", "Anna", "Peter", "Linda"],
    "Age": [24,26,22,29],
    "City": ["Paris","Berlin","New York","London"],
    "Salary": [80000,75000,70000,90000]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
data = [
    ["John",24,"Paris",80000],
    ["Anna",26,"Berlin",75000],
    ["Peter",22,"New York",70000],
    ["Linda",29,"London",90000]
]
df2 = pd.DataFrame(data)
```

```python
df2
```

```python
cols = ["Name","Age","City","Salary"]
new_df2 = pd.DataFrame(data,columns=cols,index=["a","b","c","d"])
new_df2
```

### CREATING A NEW COLUMN

```python
new_df2["Designation"] = ["Doctor","Engineer","Doctor","Engineer"]
new_df2
```

```python
# it will only return the copy but would not update the exixting
new_df2.drop("Designation",axis=1)
new_df2
```

```python
# In order to remove from the exixting we do inplace = True
new_df2.drop("Designation",axis=1, inplace=True)
new_df2
```

```python
# In order to remove 2 columns we do, we didn't do implace true as we do not want to completely remove salary age
new_df2.drop(["Salary","Age"],axis=1)
```

```python
# Removing rows is kind of same as well
new_df2.drop(["a","d"],axis=0)
```
