```python
import numpy as np 
import pandas as pd
```

### FINDING MISSING DATA

```python
# In this whole file df is always the same no changes happen as we didnot write inplace = True
data = {
    "A": [1,2,np.nan,4,5],
    "B": [1,2,3,4,5],
    "C": [1,2,3,np.nan,np.nan],
    "D": [1,np.nan,np.nan,np.nan,5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
# gives total nul values in each row
df.isna().sum()
```

```python
# checks if any null value exists in the columns
df.isna().any()
```

### REMOVING MISSING VALUES

```python
df
```

```python
# removing missing data always works on rows
df.dropna()
```

```python
df
```

```python
# threshhold: it will return the rows which have atleast n non nul values
df.dropna(thresh=3)
```

### FILLING THE MISSING DATA

```python
df
```

```python
df.fillna(10)
```

```python
# filling every column with diff values
values = {"A":10,"B":20,"C":30,"D":40}
df.fillna(value=values)
```

```python
# filling mean data
df.fillna(df.mean())
```
